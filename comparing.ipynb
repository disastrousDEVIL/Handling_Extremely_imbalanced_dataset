```python
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import numpy as np

# Example classification reports for two models
report1 = {
    "class_0": {"precision": 0.47, "recall": 1, "f1-score": 0.64},
    "class_1": {"precision": 0.001, "recall": 0.001, "f1-score": 0.001},
}

report2 = {
    "class_0": {"precision": 0.66, "recall": 0.97, "f1-score": 0.79},
    "class_1": {"precision": 0.96, "recall": 0.56, "f1-score": 0.71},
}
```

```python
# Convert classification reports to a structured format
classes = list(report1.keys())
metrics = ["precision", "recall", "f1-score"]

# Extract data for both models
data1 = [[report1[cls][metric] for metric in metrics] for cls in classes]
data2 = [[report2[cls][metric] for metric in metrics] for cls in classes]

# Convert to NumPy arrays for easier manipulation
data1 = np.array(data1)  # Shape: (n_classes, n_metrics)
data2 = np.array(data2)
```

```python
# Define class labels and metric names
x = np.arange(len(classes))  # Position for classes
width = 0.3  # Bar width for grouped bar chart

# Plot Precision, Recall, and F1-Score
for i, metric in enumerate(metrics):
    plt.figure(figsize=(8, 6))
    
    # Bar positions for two models
    pos1 = x - width / 2
    pos2 = x + width / 2
    
    # Plot bars for each model
    plt.bar(pos1, data1[:, i], width, label=f"AUG + CNN {metric.capitalize()}")
    plt.bar(pos2, data2[:, i], width, label=f"ADASYN + NCL {metric.capitalize()}")
    
    # Add labels and title
    plt.xlabel("Classes")
    plt.ylabel(metric.capitalize())
    plt.title(f"{metric.capitalize()} Comparison Across Classes")
    plt.xticks(x, classes)  # Set class names as x-ticks
    plt.legend(loc="upper right")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    plt.show()
```

```python
classes = ["Data_Augmentation + Complex CNN","ADASYN + NCL"]
metrics = ["precision", "recall", "f1-score"]
keys=['class_0',"class_1"]

# Extract data for each metric
data1 = np.array([[report1[key][metric] for metric in metrics] for key in keys])  # Model 1
data2 = np.array([[report2[key][metric] for metric in metrics] for key in keys])  # Model 2
# Define positions for bars
x = np.arange(len(classes))  # Class positions
bar_width = 0.2  # Width of each bar

# Create the plot
plt.figure(figsize=(10, 6))

colors1 = ['#4682B4', '#FFA500', '#32CD32']  
colors2 = ['#5F9EA0', '#FF8C00', '#3CB371'] 
# Plot bars for each metric and model
for i, metric in enumerate(metrics):
    plt.bar(x - bar_width + i * bar_width, data1[:, i], bar_width, label=f"AUG + CNN {metric.capitalize()}", color=colors1[i], alpha=0.8)
for i, metric in enumerate(metrics):
    plt.bar(x + i * bar_width, data2[:, i], bar_width, label=f"ADASYN + NCL {metric.capitalize()}", color=colors1[i], alpha=0.8)

# Add labels and title
plt.xlabel("Classes")
plt.ylabel("Score")
plt.title("Precision, Recall, and F1-Score Comparison")
plt.xticks(x, classes)  # Set class names as x-ticks
plt.legend(loc="upper right")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
